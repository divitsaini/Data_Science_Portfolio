# Celsius_to_Fahrenheit

## Import required libraries

```python
import tensorflow as tf
```

```python
from tensorflow import keras
```

```python
tf.__version__
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Create a sample training data

```python
temp_celsius = np.array([-40,-10,0,8,15,22,38],dtype=float)
temp_fahrenhiet = np.array([-40,14,32,46,59,72,100], dtype = float)
```

## Import models and layers

```python
from keras import models
from keras import layers
```

## Build a simple model with only one layer

```python
model = keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
model.summary()
```

OR

model = keras.Sequential()

model.add(keras.layers.Dense(units=1,input_shape=[1], name = "hidden")

model.summary()

## Compile Model

```python
model.compile(loss = "mean_squared_error", optimizer = keras.optimizers.Adam(0.1))
# 0.1 is the learning rate
```

## Fit the model/Training a model

```python
history = model.fit(temp_celsius, temp_fahrenhiet, epochs = 500)
```

## Plot model training or fitting curve

```python
plt.xlabel("Epoch Number")
plt.ylabel("Loss value")
plt.plot(history.history["loss"])
```

## Perform Predictions

```python
print(model.predict([100.0]))
```
